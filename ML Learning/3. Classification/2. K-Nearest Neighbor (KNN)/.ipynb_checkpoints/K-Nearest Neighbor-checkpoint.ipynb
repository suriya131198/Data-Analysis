# K-Nearest Neighbor (KNN)

## Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
ds = pd.read_csv(r'D:\ML Datasets\Part 3 - Classification\Section 15 - K-Nearest Neighbors (K-NN)\Python\Social_Network_Ads.csv')
x = ds.iloc[:, :-1].values
y = ds.iloc[:, -1].values
```

## Splitting the dataset into Train and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25 , random_state = 0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

## Training KNN Model in the Train dataset

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, p = 2, metric = 'minkowski')
classifier.fit(x_train, y_train)
```

```python
classifier.score(x_test, y_test)
```

## Predicting the New Result

```python
print(classifier.predict(sc.transform([[30, 87000]]))[0])
```

## Comparing the Predicted & Actual result

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
plt.matshow(cm , cmap = 'Blues', alpha = 0.75)

for i in range(2):
    for j in range(2):
        plt.text(x = j, y = i, s = cm[i,j])

plt.show()
```
