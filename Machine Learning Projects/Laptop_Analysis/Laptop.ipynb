```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\DS&DA\Dataset\laptops.csv")
```

```python
df.head()
```

```python
df.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
df.head()
```

```python
df['Brand'] = df['title'].apply(lambda x : (x.split()[0]))
```

```python
df.head()
```

```python
df['Brand'].unique()
```

```python
def Price(x):
    a = x[1:]
    z = a.split(',')
    return int(''.join(z))
```

```python
df['Price'] = df['price'].apply(lambda x : Price(x))
```

```python
df.head()
```

```python
df.drop(columns = 'price', inplace = True)
```

```python
def Discount(x):
    if '%' in x:
        a = x[1:]
        z = a.split(',')
        x = z[-1]
        Discount = x[3:]
        return int(Discount.split('%')[0])
    else:
        return("NA")
```

```python
df['Discount'] = df['discount'].apply(lambda x : Discount(str(x)))
```

```python
df.head()
```

```python
def Original(x):
    if '₹' in x:
        a = x[1:]
        z = a.split(',')
        xval = z[-1]
        Discount = xval[:3]
        z.pop()
        z.append(Discount)
        return int(''.join(z))
    elif len(x)==0:
        return('NA')
```

```python
df['Original_Price'] = df['discount'].apply(lambda x : Original(str(x)))
```

```python
df.head()
```

```python
df.groupby('Original_Price').count()
```

```python
df.head()
```

```python
def RAM(spec):
    for match in re.finditer(r"(\d+) GB/", spec):
        return int(match.group(1))
    for match in re.finditer(r"(\d+) MB/", spec):
        return int(match.group(1))/1024
    return 0
```

```python
df['RAM (GB)'] = df['title'].apply(lambda x : RAM(x))
```
