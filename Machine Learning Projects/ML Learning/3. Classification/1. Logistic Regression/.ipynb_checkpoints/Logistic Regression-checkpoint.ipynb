# Logistic Regression

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the Dataset

```python
ds = pd.read_csv(r'D:\ML Datasets\Part 3 - Classification\Section 14 - Logistic Regression\Python\Social_Network_Ads.csv')
x = ds.iloc[:, :-1].values
y = ds.iloc[:, -1].values
```

```python
ds.head()
```

```python
x
```

```python
y
```

```python
y = y.reshape(-1,1)
```

```python
y
```

## Splitting the dataset into Train set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

## Training Logistic Regression model on the Train set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

## Predicting a new result

```python
print(classifier.predict(sc.transform([[30, 87000]]))[0])
```

  ## Predicting the Testset results

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
classifier.score(x_test, y_test)
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### Confusion Matrix defining for the current problem/Dataset

    65 - Correct predictions of Customer buying an SUV
    24 - Correct Predictions of Customer not buying an SUV
    3  - Wrong prediction of Customer buying an SUV but not Bought the SUV
    8  - Wrong prediction of Customer who dodn't bought an SUV but predicted as Bought

```python
plt.matshow(cm , cmap = 'Greens' , alpha = 0.75)

for i in range(2):
    for j in range(2):
        plt.text(x = j, y = i, s= cm[i, j])

plt.show()
```
