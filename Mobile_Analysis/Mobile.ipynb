```python
#Removing data without review
df.dropna(subset = 'Review', inplace = True)
```

```python
#Seperating the Brand name from the Model coloumn and Capitalizing the data under "Brand"
df["Brand"] = df["Model"].apply(lambda x : x.split()[0].capitalize())
```

```python
df.head()
```

```python
#Getting the Color details of the Mobile
def ColorColoumn(color):
    x = color.split('(')
    val = x[-1].split(',')
    return (val[0].capitalize())
```

```python
df['Color'] = df['Model'].apply(lambda x: ColorColoumn(x))
```

```python
df.head()
```

```python
def price(mrp):
    x = mrp[1:]
    ls = x.split(',')
    res = ''.join(ls)
    return(res)
```

```python
df['MRP'] = df['Mrp'].apply(lambda x : price(x))
```

```python
df.head()
```

```python
df.drop(columns = 'Mrp', inplace = True)
```

```python
df.head()
```

```python
def review(r):
    x = r.split()
    return (x[3])
    
def rating(r):
    x = r.split()
    return (x[0])
    
def numberwc(x):
    ls = x.split(',')
    val = ''.join(ls)
    return(val)
```

```python
df['Review Count'] = df['Review'].apply(lambda x : numberwc(review(str(x))))

df['Rating Count'] = df['Review'].apply(lambda x : numberwc(rating(str(x))))
```

```python
df.head()
```

```python
df.drop(columns = 'Review', inplace = True)
```

```python
df.head()
```

```python
def RAM(x):
    if "RAM" in x:
        a = x.split("|")
        b = a[0].split()
        if "GB" in a[0]:
            return(b[0])
        elif "MB" in a[0]:
            val = int(b[0])/1024
            return(val)
        else:
            return(b[0])
    else:
        return("No Specs")
```

```python
#Splitting the RAM value into new coloumn
df['RAM(GB)'] = df['Specs'].apply(lambda x : RAM(x))
```

```python
df.head()
```

```python
df[df['RAM(GB)'] == 'No Specs'].shape
```

```python
#Filling all the available RAM details with 4 GB RAM as it is common for all
df[df['RAM(GB)'] == 'No Specs'] = 4
```

```python
df.head()
```

```python
def ROM(spec):
    for match in re.finditer(r"(\d+) GB ROM", spec):
        return int(match.group(1))
    for match in re.finditer(r"(\d+) MB ROM",spec):
        return int(match.group(1))/1024
    return 0
```

```python
df['ROM(GB)'] = df['Specs'].apply(lambda x : ROM(str(x)))
df.head()
```

```python
df['Expandable(GB)'] = df['Specs'].apply(lambda x : expandable_upto(str(x)))
df.head()
```
